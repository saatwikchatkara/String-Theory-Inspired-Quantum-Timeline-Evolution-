```python
import pennylane as qml
from pennylane import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# --- Quantum Device Setup ---
dev = qml.device("default.qubit", wires=8)

# Qubits:
# 0-2: Timeline (Branes)
# 3:   System Qubit (String)
# 4-5: Teleportation Qubits (String endpoint)
# 6-7: Ancilla for Compact Dimensions

def evolve_on_brane(system_wire, brane_id):
    """Simulate different string vibrations on each brane (timeline)."""
    if brane_id == 0:
        qml.RX(np.pi/6, wires=system_wire)
    elif brane_id == 1:
        qml.RY(np.pi/4, wires=system_wire)
    elif brane_id == 2:
        qml.RZ(np.pi/3, wires=system_wire)
    elif brane_id == 3:
        qml.RX(np.pi/2, wires=system_wire)
    elif brane_id == 4:
        qml.Hadamard(wires=system_wire)
    elif brane_id == 5:
        qml.RZ(np.pi/5, wires=system_wire)
    elif brane_id == 6:
        qml.RY(np.pi/8, wires=system_wire)
    elif brane_id == 7:
        qml.RX(np.pi/7, wires=system_wire)


def teleport_string(system, a, b):
    """Teleport the string qubit (system) using Bell pair a-b."""
    qml.Hadamard(wires=a)
    qml.CNOT(wires=[a, b])
    qml.CNOT(wires=[system, a])
    qml.Hadamard(wires=system)
    qml.CNOT(wires=[a, b])
    qml.CZ(wires=[system, b])


@qml.qnode(dev)
def string_theory_sim():
    # Setup wires
    brane = [0, 1, 2]
    system = 3
    a, b = 4, 5
    compact1, compact2 = 6, 7

    # Superposition of branes (timelines)
    for q in brane:
        qml.Hadamard(wires=q)

    # Compactified dimension entanglement (extra structure)
    qml.Hadamard(wires=compact1)
    qml.CNOT(wires=[compact1, compact2])

    # Initialize string (system qubit)
    qml.Hadamard(wires=system)

    # Evolve differently per brane
    for i in range(8):
        bin_key = format(i, '03b')
        controls = [int(k) for k in bin_key]
        qml.ctrl(lambda: evolve_on_brane(system, i), control=brane, control_values=controls)()

    # Teleport the string (string endpoint moves)
    teleport_string(system, a, b)

    return qml.state()


# --- Run Simulation ---
state = string_theory_sim()
amplitudes = state
probs = np.abs(state) ** 2
phases = np.angle(state)

# Prepare 8x8 grid for brane vs string mode
grid_amp = np.zeros((8, 8))
grid_phase = np.zeros((8, 8))

for i, amp in enumerate(amplitudes):
    if np.abs(amp) > 1e-6:
        bits = format(i, '08b')
        brane_index = int(bits[:3], 2)      # Qubits 0-2
        string_mode = int(bits[3:6], 2)     # Qubits 3-5 (includes teleportation)
        grid_amp[brane_index][string_mode] = np.abs(amp)
        grid_phase[brane_index][string_mode] = phases[i]

# --- 3D Visualization ---
X, Y = np.meshgrid(range(8), range(8))
Z = grid_amp.T
norm_phase = (grid_phase.T + np.pi) / (2 * np.pi)

fig = plt.figure(figsize=(14, 9))
ax = fig.add_subplot(111, projection='3d')
surface = ax.plot_surface(X, Y, Z, facecolors=plt.cm.twilight(norm_phase),
                          linewidth=0.3, edgecolor='k', alpha=0.95, antialiased=True)

ax.set_xlabel("Brane Index (Timeline Qubits 0–2)", labelpad=12)
ax.set_ylabel("String Mode (Qubits 3–5)", labelpad=12)
ax.set_zlabel("Excitation Amplitude", labelpad=10)
ax.set_title("String-Theory Inspired Quantum Timeline Evolution", pad=15)

fig.colorbar(surface, ax=ax, shrink=0.6, pad=0.1, label="Phase (Arg of Amplitude)")
plt.tight_layout()
plt.show()
```

```python
import pennylane as qml
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.animation import FuncAnimation

# Configuration
num_timeline_qubits = 3
num_string_qubits = 3
total_qubits = num_timeline_qubits + num_string_qubits
steps = 30  # Time steps for the animation

# Labels
timeline_labels = [f"T{i}" for i in range(2**num_timeline_qubits)]
mode_labels = [f"M{j}" for j in range(2**num_string_qubits)]

# Device
dev = qml.device("default.qubit", wires=total_qubits)

# Build time-dependent quantum circuit
def build_circuit(time):
    @qml.qnode(dev)
    def circuit():
        for i in range(num_timeline_qubits):
            qml.Hadamard(wires=i)
            qml.RZ(time * np.pi / 8, wires=i)  # Add evolving phase shift

        for i in range(num_timeline_qubits):
            qml.CNOT(wires=[i, i + num_timeline_qubits])

        for i in range(total_qubits):
            qml.PhaseShift(time * np.pi / 12 * (i+1), wires=i)  # Time-evolving phase

        return qml.state()
    return circuit

# Generate evolving probability matrices
prob_matrices = []

for t in range(steps):
    state = build_circuit(t)()
    state_matrix = state.reshape((2**num_timeline_qubits, 2**num_string_qubits))
    prob_matrix = np.abs(state_matrix)**2
    prob_matrix /= prob_matrix.sum(axis=1, keepdims=True)
    prob_matrices.append(prob_matrix)

# --- Animation using matplotlib ---
fig, ax = plt.subplots(figsize=(8, 6))
cmap = "viridis"
sns.set(style="whitegrid")

def update(frame):
    ax.clear()
    sns.heatmap(prob_matrices[frame], annot=False, fmt=".2f", cmap=cmap,
                xticklabels=mode_labels, yticklabels=timeline_labels, cbar_kws={"label": "P(i → j)"}, ax=ax)
    ax.set_title(f"🌀 Quantum Teleportation Matrix — Time Step {frame}")
    ax.set_xlabel("String Modes")
    ax.set_ylabel("Timelines")

ani = FuncAnimation(fig, update, frames=steps, interval=300)

# Show animation
plt.tight_layout()
plt.show()

# Optional: Save animation as GIF or MP4
# ani.save("quantum_teleportation_timelines.gif", dpi=150, writer="pillow")
```
